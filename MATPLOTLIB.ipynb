# 5) BOXPLOT

1.minimum  2.maximum  3.Q1 4.Q3  5.Q2=median

```python
d=np.array([0,21,31,51,61,71,110,114,116,191,213])
```

```python
d
```

```python
median=np.median(d)
median
```

```python
quartile=np.percentile(d,[25,50,75])
quartile
print('Q1: %.3f' %quartile[0])
print('Q2: %.3f' %quartile[1])
print('Q3: %.3f' %quartile[2])
```

```python
minimum=min(d)
maximum=max(d)
print('max: %.3f'%minimum )
print('min: %.3f'%maximum )
```

```python
style.use('ggplot')
plt.boxplot(d)
```

# 6) PIE CHART

```python
labels=['ML','ds','teac','cc']
sized=[245,130,215,210]
colors=['pink','blue','yellow','green']
explode=(0.3,0,0,0)                 #the gap present between ML and ds,cc
plt.pie(sized,explode=explode,labels=labels,colors=colors,autopct='%.2f%%',shadow=False)
plt.axis('equal')            #360 degree equally
```

# 7) Histogram

```python
import random         #for generating random number, not very imp for histogram its only for eg.
```

```python
age1=np.random.randint(18,45,(100))
age2=np.random.randint(15,40,(100))
age1
```

```python
plt.hist(age1)
plt.title("age1 histogram")
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(age1,bins,color='yellow',rwidth=0.9)                    #bins is useful in many aspect
plt.title("age1 histogram")
```

width orientation

```python
bins=[15,20,25,30,35,40,45]
plt.hist(age1,bins,rwidth=0.9,orientation='horizontal',color='g')         #bins is useful in many aspect
plt.title("age1 histogram")
```

```python
plt.hist(age1,bins,rwidth=0.9,color='g',label='age1') 
plt.hist(age2,bins,rwidth=0.9,color='y',label='age2')
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('no of student')
plt.legend()
plt.figure(figsize=(5,5))
```

in above histogram in first 3 columns age1 is hide behind age2 so for showing both downward code is there

```python
plt.hist([age1,age2],bins,rwidth=0.9,color=['b','g'],label=['1','2'])
plt.title('age distribution')
plt.xlabel('age')
plt.ylabel('no of student')
plt.legend()
plt.figure(figsize=(10,10))
```
